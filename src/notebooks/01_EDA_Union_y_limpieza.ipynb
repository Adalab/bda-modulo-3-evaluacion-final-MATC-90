### Merging the DataFrames

```python
df_merge = df_flight_activity.merge(df_loyalty, on= ['Loyalty Number'], how = 'right')
```

```python
df_merge.head()
```

### Data Cleaning

**Data Cleaning Checklist:**

* ⬜ Create new columns with customer's status.
* ⬜ Drop 'Points Redeemed', 'Dollar Cost', 'Cancellation Year' and 'Cancellation Month' columns from the DataFrame.
* ⬜ Remove the duplicate records from the DataFrame.
* ⬜ Clean the negative numbers from 'Salary' column.
* ⬜ Handle null values in the 'Salary' column.

```python
# Create a new column indicating whether a customer is active or not by handling the null values.
df_merge['Customer Active'] = df_merge['Cancellation Year'].isnull()
```

```python
# Dropping columns that will not be used in the study.
df_merge.drop(columns=['Dollar Cost Points Redeemed','Points Redeemed','Cancellation Year','Cancellation Month'], inplace=True)
```
