## 03. Data Cleaning

**Data Cleaning Checklist:**

* ⬜ Create new columns with customer's status.
* ⬜ Drop `Points Redeemed`, `Dollar Cost`, `Cancellation Year` and `Cancellation Month` columns from the DataFrame.
* ⬜ Remove the duplicate records from the DataFrame.
* ⬜ Clean the negative numbers from `Salary` column.
* ⬜ Handle null values in the `Salary` column.

```python
# Create a new column indicating whether a customer is active or not by handling the null values.
df_merge['Customer Active'] = df_merge['Cancellation Year'].isnull()
```

```python
# Dropping columns that will not be used in the study.
df_merge.drop(columns=['Dollar Cost Points Redeemed','Points Redeemed','Cancellation Year','Cancellation Month'], inplace=True)
```

```python
df_merge.duplicated().sum()
```

```python
# Dropping duplicated rows
df_merge.drop_duplicates(inplace=True)
```

```python
df_merge.duplicated().sum()
```

```python
# The presence of negative values in the 'Salary' column is nonsensical. 
# Since negative numbers are not repeated, it is assumed to be a typo; therefore, it was decided to convert these values to positive numbers.

df_merge['Salary'] = df_merge['Salary'].abs()
```

```python
df_merge['Salary'].min()
```

```python
# The missing value percentage remains identical after merging. 
# Since it exceeds 20%, advanced imputation techniques will be applied.
round(df_merge['Salary'].isnull().sum() / len(df_merge) * 100, 2)
```

```python
# Set up the IterativeImputer with 10 cycles and random state 42
imputer_iter = IterativeImputer(max_iter= 10, random_state= 42)

# Select all numerical columns to serve as predictors for the model
merge_numerical_cols = df_merge.select_dtypes(include=['number']).columns

# Fit and transform the numerical columns (returns a NumPy array)
imputed_values = imputer_iter.fit_transform(df_merge[merge_numerical_cols])

# Convert the resulting array back into a temporary DataFrame
df_temp = pd.DataFrame(imputed_values, columns=merge_numerical_cols, index=df_merge.index)

# Safely overwrite ONLY the 'Salary' column in our main DataFrame
df_merge['Salary'] = df_temp['Salary']

print(f"Remaining missing values in 'Salary': {df_merge['Salary'].isnull().sum()}")
```

```python
# Verify that all data types are correct and missing values have been resolved
df_merge.info()
```
